```python
import numpy as np
import matplotlib.pyplot as plt
```

**Generate data using dummy function**

```python
X = np.arange(10)
Y = (X-5)**2 + 3

print(X, Y)
```

**Visualize the data**

```python
plt.style.use("seaborn-v0_8-whitegrid")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**Gradient Descent**

$$ x = x - lr * \frac{\partial {x}}{\partial {f}{x}} f(x)$$

```python
def gradient_descent(init, learning_rate, _X, _Y):
    # 1. Initialize x with a random value
    x = init
    # 2. Set learning rate value
    lr = learning_rate

    plt.plot(_X, _Y)
    
    # 3. Get the minimum value of the curve
    for i in range(50):
        grad = 2*(x-5) # derivative of Y = (X-5)**2 + 3
        x = x - lr * grad
        y = (x-5)**2 + 3
        plt.scatter(x, y)

    plt.show()
```

```python
gradient_descent(init=0, learning_rate=0.1, _X=X, _Y=Y)
```

**What happend if the learning rate is bigger?**

The number of steps to reach the minimum value are less. So, it learn faster.

```python
gradient_descent(init=0, learning_rate=0.5, _X=X, _Y=Y)
```

**But, What if the learning rate is too big?**

The algorithm diverges. It could be possible the algorithm never converge to the minimum value.

```python
gradient_descent(init=0, learning_rate=0.9, _X=X, _Y=Y)
```
